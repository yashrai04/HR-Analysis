# Feature Engineering and Selection

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encoder = LabelEncoder()
df1['department'] = label_encoder.fit_transform(df1['department'])
df1['education'] = label_encoder.fit_transform(df1['education'])
df1['gender'] = label_encoder.fit_transform(df1['gender'])
df1['recruitment_channel'] = label_encoder.fit_transform(df1['recruitment_channel'])
```

```python
#Calculate the mean promotion rate per department
df1['department_promotion'] = df1.groupby('department')['is_promoted'].transform('mean')

#Calculate the average age of promoted employees per department
df1['promotion_age_department'] = df1[df1['is_promoted'] == 1].groupby('department')['age'].transform('mean')

# Calculate the number of promotions per education level
df1['education_promotiom'] = df1.groupby('education')['is_promoted'].transform('sum')

# Calculate the percentage of promotions per gender
df1['gender_promotion'] = (df1.groupby('gender')['is_promoted'].transform('sum') / df1['is_promoted'].sum()) * 100

# Calculate the average length of service for promoted employees per department
df1['promoted_department'] = df1[df1['is_promoted'] == 1].groupby('department')['length_of_service'].transform('mean')
```

```python
df1
```

```python
region_onehot = pd.get_dummies(df1['region'], prefix='region', drop_first=True)
df1= pd.concat([df1, region_onehot], axis=1)
df1.drop('region', axis=1, inplace=True)
```

```python
correlation_matrix = df1.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, annot_kws={'size': 4})
plt.title("Correlation Matrix")
plt.show()
```
